```python
paths = {
    'online_retail': '../data/online_retail.csv',
    'profiling_table': '../outputs/profiling_table.csv'
}
```

```python
# import dataset
df = pd.read_csv(paths['online_retail'])
```

```python
# inspect dataset
df.head(3)
```

## Part A Column profiling table

```python
# create profile table
summary = pd.DataFrame({
    "col_name": df.columns,
    "dtype": df.dtypes,
    "n_missing": [df[col].isna().sum() for col in df.columns],
    "missing_rate": [round(df[col].isna().mean()*100, 2) for col in df.columns],
    "is_unique": df.columns.duplicated(),
    "n_unique": [df[col].nunique() for col in df.columns],
    "sample_values": [list(df[col].unique()[0:3]) for col in df.columns]
})
summary.to_csv(paths['profiling_table'], index=False)
summary
```

## Part B Short Questions

#### 1. rows/cols numbers 

```python
print(f'{df.shape[0]} is the number of rows')
print(f'{df.shape[1]} is the number of columns')
```

#### 2. top 3 columns by missing rate

```python
missing_rates = {}
for col in df.columns:
    missing_rate = df[col].isna().mean()
    missing_rates[col] = missing_rate
top3missing = sorted(missing_rates.items(), key = lambda x: x[1], reverse = True)[:3]
top3missing
print(top3missing)
```

#### 3. top 5 countries by count

```python
top5countries = df.groupby('Country').count().index[:5]
print(top5countries)
```

#### 4. number of unique InvoiceNo

#### 5. missing rate of CustomerID

```python
print(f'{round(df['CustomerID'].isna().mean() * 100 , 2)}% is the missing rate of CustomerID')
```

```python
print(f'{sum(df['Quantity']<= 0)} is number of rows with Quantity <= 0')
print(f'{sum(df['UnitPrice']<= 0)} is number of rows with Quantity <= 0')
```
